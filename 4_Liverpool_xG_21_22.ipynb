```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/Liverpool xG.csv')
df
```

```python
# Dropping all missing xG and xA values
# Considering only PL and UCL matches

df = df.dropna()
df.reset_index(drop = True)
```

```python
x_axis = np.arange(len(df['G-xG']))

plt.bar(x_axis - 0.1, df['G-xG'], width = 0.2, label = 'Goals - xG')
plt.bar(x_axis + 0.1, df['A-xA'], width = 0.2, label = 'Assists - xA')

#plt.xticks(x_axis, df['Round'])
plt.legend()

plt.show()
```
