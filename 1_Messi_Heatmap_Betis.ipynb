```python
from mplsoccer.pitch import Pitch
```

```python
df = pd.read_csv('/content/Messi v Betis.csv')
```

```python
# Converting data to match statsbomb pitch
df['x'] = df['x'] * 1.2
df['y'] = df['y'] * .8
df['endX'] = df['endX'] * 1.2
df['endY'] = df['endY'] * .8
df
```

```python
fig, ax = plt.subplots(figsize = (13.5, 8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')
Pitch()
pitch = Pitch(pitch_type = 'statsbomb', pitch_color = '#22312b')

pitch.draw(ax = ax)
plt.gca().invert_yaxis()

kde = sns.kdeplot(
        df['x'],
        df['y'],
        shade = True,
        shade_lowest=False,
        alpha=.5,
        n_levels=10,
        cmap = 'magma'
)

plt.title('Messi HeatMap vs Real Betis', size = 20, color = 'white')
plt.xlim(0,120)
plt.ylim(0,80)
```
