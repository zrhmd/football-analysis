```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
df = pd.read_csv('/content/Player stats.csv')
df
```

```python
# Filtering out all the defenders

df = df[df['Pos'] == 'DF']
df.reset_index()
```

```python
#Considering players that have played atleast 15 games

df = df[df['90s'] > 15]
df.reset_index()
```

```python
vvd_stats = df[df['Player'] == 'Virgil van Dijk']
vvd_stats
```

```python
plt.scatter(df['TklW'], df['Won%'], s=1, c='blue')
plt.scatter(vvd_stats['TklW'], vvd_stats['Won%'], s = 15, c = 'red')
plt.xlabel('Tackles won')
plt.ylabel('Aerial Duels %')
plt.title('Comparison of Europe\'s Top 5 League\'s Defenders')
plt.text(5,80,'Virgil van Dijk')
plt.show()
```
